```python
# single request

from time import sleep
from ebaysdk.finding import Connection

def item_request(p):

        if __name__ == '__main__':
            api = Connection(config_file='ebay.yaml', debug=True, siteid="EBAY-US")

            request = {
                'categoryId': ['1281'],
                'itemFilter': [
                    {'name': 'condition', 'value': 'new'}
                ],
                'paginationInput': {
                    'entriesPerPage': 100,
                    'pageNumber': p
                },
                'sortOrder': 'StartTimeNewest'
            }

            response = api.execute('findItemsByCategory', request)
            resanswer = response.json()
            jload = json.loads(resanswer)
            sr = jload['searchResult']
            srlist = sr['item']
        
            itemdf = pd.DataFrame(columns=['categoryId', 'itemId', 'categoryName', 'title', 'Price', 'galleryURL',
                                 'viewItemURL'])
            for item in srlist:
                row = {
                    'itemId': item.get('itemId'),
                    'title': item.get('title'),
                    'categoryId': item.get('primaryCategory').get('categoryId'),
                    'categoryName': item.get('primaryCategory').get('categoryName'),
                    'Price': item.get('sellingStatus').get('currentPrice').get('value'),
                    'galleryURL' : item.get('galleryURL'),
                    'viewItemURL': item.get('viewItemURL')
                        }
                itemdf = itemdf.append(row, ignore_index=True)
               
        return itemdf
```

```python
itemdf = []
for i in range(1, 16):
    itemdf.append(item_request(i))
```

```python
df = pd.concat(itemdf, axis = 0)
```
