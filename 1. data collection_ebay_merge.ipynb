```python
import glob
```

```python
csv_files = glob.glob('data/filt/*.{}'.format('csv'))
csv_files
```

```python
df_merged = pd.DataFrame()
 
# append the CSV files
for file in csv_files:
    df = pd.read_csv(file)
    df_merged = df_merged.append(df, ignore_index=True)
 
df_merged
```
